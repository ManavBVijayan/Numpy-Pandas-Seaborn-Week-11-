#### Local Mean Interpolation

```python
import pandas as pd
import numpy as np
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5, 6, np.nan, 8, 9],
    'B': [np.nan, 22, 23, np.nan, 25, 26, 27, 28, np.nan]
}
data=pd.DataFrame(data)
```

```python
data

```

```python
n_size=4
for col in data.columns:
    for i in range(len(data[col])):
        if pd.isna(data.at[i,col]):
            left=max(0,i-n_size)
            right=min(i+n_size,len(data[col])-1)
            neighbours=data[col][left:right+1]
            neighbours=neighbours.dropna()
            if len(neighbours)>0:
                data.at[i,col]=neighbours.mean()
data
```

#### Global Mean Interpolation

```python
data = {
    'A': [1, 2, np.nan, 4, 5, 6, np.nan, 8, 9],
    'B': [np.nan, 22, 23, np.nan, 25, 26, 27, 28, 29]
}
```

```python
data=pd.DataFrame(data)
```

```python
for col in data.columns:
    mean_value=data[col].mean()
    for i in range(len(data[col])):
        if pd.isna(data.at[i,col]):
            data.at[i,col]=mean_value
```
