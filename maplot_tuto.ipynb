## FILL_BETWEEN PLOT

```python
#data
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
```

```python
# create fill_between plt
plt.plot(x,y1,color="g",label="Sin(x)")
plt.plot(x,y2,color="r",label="Cos(x)")
```

```python
plt.fill_between(x,y1,y2 ,where=(y1>y2), interpolate=True,alpha=0.3, color='gray', label='Filled Area')
plt.show()
```

```python
plt.style.use('_mpl-gallery')

# make data
np.random.seed(1)
x = np.linspace(0, 8, 16)
y1 = 3 + 4*x/8 + np.random.uniform(0.0, 0.5, len(x))
y2 = 1 + 2*x/8 + np.random.uniform(0.0, 0.5, len(x))

# plot
fig, ax = plt.subplots()

ax.fill_between(x, y1, y2, alpha=.5, linewidth=0)
ax.plot(x, (y1 + y2)/2, linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

## STACKPLOT

```python
# data
days=[1,2,3,4,5,6,7]
task1=[3,4,1,6,9,10,16]
task2=[6,9,10,13,2,15,7]
task3=[2,12,16,19,20,5,6]
```

```python
plt.figure(figsize=(15,10))
plt.stackplot(days,task1,task2,task3, labels=["task1","task2","task3"])
plt.xlabel("days")
plt.ylabel("tasks")
plt.title("Task completion over time")
plt.legend(loc='upper left')
plt.show()
```

## ERRORBAR

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])
y_err = np.array([0.2, 0.3, 0.1, 2, 2])  
plt.errorbar(x, y, yerr=y_err, fmt='o', color='b', label='Data with Error Bars')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Error Bar Plot Example')
plt.legend()
plt.show()
```

### PIE CHART

```python
labels = ['A', 'B', 'C', 'D']
sizes = [30, 45, 15, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # Explode the 1st slice (i.e., 'A')
plt.figure(figsize=(6, 6)) 
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Pie Chart Example')

plt.show()
```

```python
categories = ['Start', 'Category 1', 'Category 2', 'Category 3', 'End']
values = [0, 20, -15, 30, 35]

# Calculate cumulative values
cumulative_values = [sum(values[:i+1]) for i in range(len(values))]
plt.figure(figsize=(15,10))

# Create a waterfall chart
plt.bar(categories, cumulative_values, color='b', align='center')
plt.bar(categories, values, color='g', align='center')

# Customize labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Waterfall Chart Example')

# Show the plot
plt.show()
```
